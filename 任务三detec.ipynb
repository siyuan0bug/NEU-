```python
from pathlib import Path
from ultralytics import YOLO

# 加载预训练的YOLOv8n模型
model = YOLO('/kaggle/input/yolopt/best1.pt')

# 推理一个文件夹里的所有图片
folder_path = '/kaggle/input/yoloseg/test'  # 请替换为实际的文件夹路径

# 获取文件夹中所有图片的路径
image_paths = [str(img_path) for img_path in Path(folder_path).rglob("*.jpg")]

# 进行推理并查看结果
for image_path in image_paths:
    results = model(image_path)
    for r in results:
        print(r.boxes)  # 打印包含检测边界框的Boxes对象
    

```
